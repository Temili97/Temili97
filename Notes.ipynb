```python
def animal_acceptance(data_animals):
    if data_animals['furniture']=="not furnished" and data_animals["Animal_Acceptance"]=="not acept":
        return "Not interested"
X = house_data[house_data['House_ID'] == 344].iloc[0]

print(animal_acceptance(X))
```

```python
def checking(i):
    if np.where(house_data[house_data["House_ID"]==i][["Animal_Acceptance"]]=="not accept", True, False) and np.where(house_data[house_data["House_ID"]==i]["furniture"]=="not furnished", True, False):
        return "not interested"
    else:
        return "interested"
checking(344)
```

```python
def animal_info(i):
    if i['furniture']=="not furnished" and i['Animal_Acceptance']=="not acept":
        return "Not interested"
    
x = house_data[house_data['House_ID']==344].iloc[0]

print(animal_info(x))
```

```python
ticker=['FB']
```

```python
data_stocks=data.DataReader(name= ticker,
                           data_source='yahoo',
                           start='2020-01-01',
                           end='2021-01-01')
```

```python
fig=go.Figure()

fig.add_trace(go.Scatter(x=data_stocks.index,
                        y=data_stocks.Open,
                        mode='lines',
                        name='Open Price of FB'))

fig.add_trace(go.Scatter(x=data_stocks.index,
                        y=data_stocks.Close,
                        mode='lines',
                        name='Close Price of FB'))


fig.update_layout(title='Comparison of Open and Close Prices of FB')

fig.show()
```
